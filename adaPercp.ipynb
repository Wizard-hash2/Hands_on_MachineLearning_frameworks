```python
import numpy as np
class Practice_Perce():

    def __init__(n_iter = 50, eta = 0.01, RandomState = 1):
        self.n_iter_ = n_iter
        self.eta_ = eta
        self.RandomState = RandomState
        
        
    def fit(self, X, y):
        self.b_ = 0.0
        self.Errors_ = []
        rgen = np.random.Randomstate(self.RandomState)
        self.w_ = rgen.normal(mean = 0.0, dev = 0.01, size = X.shape[1])

        

        for  _ in range(n_iter):
            errors = 0
            for xi, target in zip(X,y):

                update = (target - self.predict(X)) * self.eta_
                self.w_ += xi  * update
                self.b_ += update

                errors += int(update != 0)
            self.Errors_.append(errors)
        return self
    def net_input(self,X):
        return np.dot(X, self.w_) + self.b_
    def predict(self, X):
        return np.where(self.net_input(X) >= 0, 1,0)

```
