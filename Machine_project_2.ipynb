```python
%matplotlib inline 
#render static visualization below the
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)

plt.plot(x,np.sin(x))
```

```python
plt.xlim([2,8])
plt.ylim([0.25,0.75])
plt.plot(x,np.sin(x))
plt.show()
```

```python
x = np.linspace(0,10,10)
plt.plot(x,np.sin(x),label = ("Sinx"), linestyle = "", marker="o",)
plt.xlabel("x")
plt.ylabel("fx")
plt.legend(loc="lower left")
```

```python
RNG = np.random.RandomState(123)

x = RNG.normal(size = 500)
y = RNG.normal(size = 500)
plt.scatter(x,y)

```

```python
means = [5,8,10]

std_dev = [0.2,0.4,0.6]

x_labels = ["NO_1","No_2","NO_3"]

y = list(range(len(x_labels)))
#Determint no. of x labels and then convert it to range and then change them to list

plt.bar(y,means, yerr = std_dev)
```

```python
rng = np.random.RandomState(123)
x = rng.normal(0,20,1000)

bin = np.arange(-100,100,5)

plt.hist(x,bins = bin)


```

```python
rng = np.random.RandomState(123)
#same number of random number
x1 = rng.normal(0,20,1000) 
#0 mean 20 std dev 10000 values
x2 = rng.normal(15,10,1000)

bin = np.arange(-100,100,5)

plt.hist(x1, bins = bin, alpha= 0.5)
plt.hist(x2, bins = bin, alpha= 0.5)
```

```python
x = range(10)
#rint(np.random.RandomState(46))
y = range(10)

fig, ax = plt.subplots(nrows = 2, ncols = 5, sharex = True,sharey = True )

for val in ax:
    for row in val:
        row.plot(x,y)

```

```python
x = np.linspace(0,10,100,endpoint=False)
plt.xlim(2,8)
plt.xlabel("x")

plt.plot(x , np.sin(x),color = "blue",marker = "o",linestyle = " ",label = "SInx")
plt.legend(loc = "upper left")
plt.savefig("My_Plot_Example.pdf")
plt.show()
```
